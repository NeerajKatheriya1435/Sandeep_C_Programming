# Pandas Series Lecture 1

```python
import pandas as pd

data1=pd.Series([5,3,12,18])
print(data1)
print(type(data1))
```

### Data with index

```python
data2=pd.Series([5,3,2,8],index=["A","B","C","D"],dtype=float,name="Python Data")
print(data2)

# {"Name":"Shivam","Age":67}
```

### Data From Dictionary

```python
dic={"std":["phy","chem","math"],"marks":[20,45,89],"rank":[2,5,7]}
ser1=pd.Series(dic)
print(ser1)
print(type(ser1))
```

```python
dic1={1:[34,23,23],2:[20,45,89],3:[2,5,7]}
ser2=pd.Series(dic1)
print(ser2)

# tup1=(6,7,5)
# print(type(tup1))
```

### Single Data with manual index

```python
s1=pd.Series(12,index=[1,3,5,7])
print(s1)
print(type(s1))
```

```python
s1=pd.Series([6,4,3,4],index=[1,2,3,4])
# print(s1)
s2=pd.Series([3,6,1,8],index=[1,2,3,4])
# print(s2)
print(s1-s2)
```

```python
s1=pd.Series(12,index=[1,2,5,7,8,9])
s2=pd.Series(12,index=[1,3,5,7])
print(s1+s2)
```

### Data Frame

```python
l1={"a":[1,2,4,6],"s":[4,6,8,9],"d":[2,3,5,6]}
data1=pd.DataFrame(l1)
print(data1)
print(type(data1))
```

### Specific Column View

```python
p2={"a":[1,2,4,6],"s":[4,6,8,9],"d":[2,3,5,6]}
data2=pd.DataFrame(p2,columns=["a","s"],index=[1,3,4,5])
print(data2)
print(type(data2))
```

```python
list1=[[2,3,4],[3,5,6],[2,8,4]]
d2=pd.DataFrame(list1)
print(d2)
```

```python
s1={"a":pd.Series([4,2,9,6]),"b":pd.Series([2,9,5,3])}
data2=pd.DataFrame(s1)
print(data2)
# Access the element
print("Data:",data2["a"][1])
```

### Arithmetic operation in DataFrame

```python
import pandas as pd 
d4=pd.DataFrame({"a":[1,2,4,6],"s":[4,6,8,9],"d":[2,3,5,6]})
# print(d4)
d4["sum"]=d4["a"]+d4["s"]
d4
```

```python
d4["Sum"]=d4["a"]+d4["s"]+d4["d"]
d4
d4["Diff(a-s)"]=d4["a"]-d4["s"]
d4
d4["Mul(a*s)"]=d4["a"]*d4["s"]
d4
d4["Div(a/s)"]=d4["a"]/d4["s"]
d4
```

```python
d4["Sum>20"]=d4["Sum"]>=10
d4
```

### Insert and Delete Data

```python
l1={"a":[1,2,4,6],"s":[4,6,8,9],"d":[2,3,5,6]}
data1=pd.DataFrame(l1)
data1.insert(1,"Data1",[7,5,3,0])
print(data1)
```

```python
l1={"a":[1,2,4,6],"s":[4,6,8,9],"d":[2,3,5,6]}
data1=pd.DataFrame(l1)
data1.insert(3,"Data1",data1["s"])
print(data1)
```
